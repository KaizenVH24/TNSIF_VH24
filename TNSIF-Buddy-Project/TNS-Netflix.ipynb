```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Dataset
df = pd.read_csv(r'C:\Users\vinay\Downloads\netflix_titles.csv')
```

```python
df.head()
```

Step 3

```python
# Data Cleaning - Missing Values
df.isnull().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df.dropna(subset=['date_added', 'rating'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# 1. Plot Content Type Distribution

plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=df)
plt.title('Content Type Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Top 10 Directors by Number of Titles

top_directors = df[df['director'] != 'Unknown']['director'].value_counts().nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='viridis')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

```python
df['duration'].fillna('0', inplace=True)
```

```python
# 3. Average Duration of Movies and TV Shows

movie_duration = df[df['type'] == 'Movie']['duration'].str.replace(' min', '').astype(int).mean()
tv_show_seasons = df[df['type'] == 'TV Show']['duration'].str.replace(' Season(s)?', '', regex=True).astype(int).mean()

print(f"Average Movie Duration: {movie_duration:.2f} minutes")
print(f"Average TV Show Seasons: {tv_show_seasons:.2f} seasons")
```

```python
# 4. Most Common Genres

top_genres = df['listed_in'].str.split(', ').explode().value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='mako')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

```python
# 5. Content Added Over Time

# Clean and convert the 'date_added' column to datetime objects
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

# Drop rows where 'date_added' could not be parsed
df.dropna(subset=['date_added'], inplace=True)

# Extract the year from the 'date_added' column
df['year_added'] = df['date_added'].dt.year

# Get the count of content added per year and sort it by year
content_added_over_time = df['year_added'].value_counts().sort_index()

# Plot the results
plt.figure(figsize=(12, 6))
content_added_over_time.plot(kind='line', marker='o')
plt.title('Content Added to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.show()
```

```python
# 6. Ratings Distribution

plt.figure(figsize=(12, 6))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title('Distribution of Content Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
# 7. Country-wise Content Analysis

top_countries = df[df['country'] != 'Unknown']['country'].str.split(', ').explode().value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='rocket')
plt.title('Top 10 Content Producing Countries on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
# 8. Trend Analysis for a Genre (e.g., 'Dramas')

# Filter the DataFrame for titles that include 'Dramas' in their genres
dramas_df = df[df['listed_in'].str.contains('Dramas', na=False)]

# Get the count of dramas added per year and sort it by year
dramas_over_time = dramas_df['year_added'].value_counts().sort_index()

# Plot the trend for Dramas
plt.figure(figsize=(12, 6))
dramas_over_time.plot(kind='line', marker='o')
plt.title('Trend of Dramas Added to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Dramas Added')
plt.grid(True)
plt.show()
```

```python
# 9. Duration vs Ratings for Movies

movies_df = df[df['type'] == 'Movie'].copy()
movies_df['duration'] = movies_df['duration'].str.replace(' min', '').astype(int)

plt.figure(figsize=(14, 8))
sns.boxplot(x='rating', y='duration', data=movies_df, order=movies_df['rating'].value_counts().index)
plt.title('Movie Duration vs. Rating')
plt.xlabel('Rating')
plt.ylabel('Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
```
